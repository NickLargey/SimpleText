```python
import os
import huggingface
import transformers
from transformers import AutoTokenizer
import torch
import pandas as pd
import nltk
nltk.download('punkt')


if torch.cuda.is_available():
    torch.set_default_device("cuda")
    print("cuda")
else:
    torch.set_default_device("cpu")
model = "meta-llama/Llama-2-7b-chat-hf"
# model = "meta-llama/Llama-2-13b-chat-hf"

tokenizer = AutoTokenizer.from_pretrained(model)
pipeline = transformers.pipeline(
    "text-generation",
    model=model,
    torch_dtype=torch.float16,
    device_map="auto",
    max_new_tokens=7500,
)


system_message = ["You are Yan Lecun", "You have been a software engineer for 25 years working for a major company like IBM", "You are a child of above average intellegence.", "You are a Community College Professor.", "You are your creators"]
user_message = ["You want to insert n items in a queue, and then remove them one by one. However, the only data structure that you can use is Stack. How would you do it? Just explain your approach.",
"Between selection sort and insertion sort, which one is preferred and why?",
# A: Insertion sort, as its best-case, is O(N) vs. O(N2) for selection sort
"Classes can implement interface X, which has a compare/compareTo method. What is X? [Just the name]",
# A: Comparator, Comparable
"What Java keyword is used when creating a new class based on a parent class?",
# A: extends
"Which tree traversal is used when deleting a tree?"]
# A: Pre-order

for i in range(5):
  prompt = "<s>[INST]<<SYS>>" + system_message[i] + \
      "<</SYS>>\n" + user_message[i] + "[/INST]</s>"
  sequences = pipeline(
      prompt,
      do_sample=True,
      top_k=1,
      num_return_sequences=1,
      eos_token_id=tokenizer.eos_token_id
  )
  # print(len(sequences))
  for seq in sequences:
      print(seq)
      result = seq['generated_text']
```
